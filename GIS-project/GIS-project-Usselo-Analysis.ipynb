```python
import os

import rasterio
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
from collections import OrderedDict
```

```python
site = "Usselo-Enschede/usselo-1"
font = {'size'   : 10}

matplotlib.rc('font', **font)

images_ndvi = OrderedDict()
images_rgb = []
for file in sorted(os.listdir(site)):
    file_name = os.path.join(site, file)
    with rasterio.open(file_name, 'r') as ds:
        img = ds.read()
    img = np.transpose(img, [1, 2, 0])
    img_ndvi = img[:, :, [23]].flatten()
    img_rgb = img[:, :, [3, 2, 1]]
    img_ndvi = img_ndvi[np.nonzero(img_ndvi)]
    name = file[:-4]
    images_ndvi[name] = img_ndvi
    images_rgb.append(img_rgb)
fig, axes = plt.subplots(len(images_ndvi), 1, figsize=(6, 12))
for i in range(len(images_ndvi)):
    sns.histplot(list(images_ndvi.items())[i], kde=True, ax=axes[i], binwidth=0.05, binrange=[0, 1])
    axes[i].set_title("Hilgenhuesweg, " + list(images_ndvi.items())[i][0])
    axes[i].set_xlim([0, 1])
    axes[i].set_ylim([0, 40])

```
