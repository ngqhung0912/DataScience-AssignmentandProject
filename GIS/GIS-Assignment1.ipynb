```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
```

```python
with rasterio.open("Datasets/R_34FN2.TIF") as dataset:
    geo_data = dataset.read()[0]
geo_data
```

```python
plt.figure()
plt.imshow(geo_data, cmap="gist_stern")
```

```python
enschede_center = geo_data[6900:7900, 5700:6700] #row, col
```

```python
enschede_masked = np.ma.masked_where(enschede_center > np.median(enschede_center), enschede_center)
```

```python
plt.imshow(enschede_masked)
```
